```python
import pandas as pd

df = pd.read_csv("DiseaseAndSymptoms.csv")

print("Rows & Columns:", df.shape)
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
# Identify all symptom columns
symptom_cols = [col for col in df.columns if "Symptom" in col]

# Combine all symptom columns into one string
df["Symptoms"] = df[symptom_cols].apply(
    lambda row: ", ".join([str(x) for x in row.dropna().tolist() if str(x).strip() != ""]),
    axis=1
)

# Clean: remove extra spaces
df["Symptoms"] = df["Symptoms"].str.strip().str.lower()

df[["Disease", "Symptoms"]].head()
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.20, random_state=42)

print("Train rows:", len(train_df))
print("Test rows:", len(test_df))
```

```python
import json

def make_entry(symptoms, disease):
    return {
        "instruction": "Identify the disease pattern based on symptoms.",
        "input": symptoms,
        "output": f"Disease: {disease}\nExplanation: Symptoms match patterns of {disease} in the dataset.\nNote: This is not medical advice."
    }
```

```python
with open("train.jsonl", "w") as f:
    for _, row in train_df.iterrows():
        entry = make_entry(row["Symptoms"], row["Disease"])
        f.write(json.dumps(entry) + "\n")

print("train.jsonl created!")
```

```python
with open("test.jsonl", "w") as f:
    for _, row in test_df.iterrows():
        entry = make_entry(row["Symptoms"], row["Disease"])
        f.write(json.dumps(entry) + "\n")

print("test.jsonl created!")
```

```python
import pandas as pd

df = pd.read_csv("DiseaseAndSymptoms.csv")
df.head()
```

```python
def convert_row(row):
    symptoms = ", ".join([str(row[c]) for c in row.index if "Symptom" in c and row[c] != 0])
    return f"Symptoms: {symptoms}. Disease: {row['Disease']}"

df["text"] = df.apply(convert_row, axis=1)
df["text"].head()
```

```python
from sklearn.model_selection import train_test_split

train_texts, val_texts = train_test_split(df["text"], test_size=0.1, random_state=42)
```

```python
tokenizer.pad_token = tokenizer.eos_token
train_encodings = tokenizer(list(train_texts), truncation=True, padding=True)
val_encodings = tokenizer(list(val_texts), truncation=True, padding=True)
```

```python
import torch

class DiseaseDataset(torch.utils.data.Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __len__(self):
        return len(self.encodings["input_ids"])

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item["labels"] = item["input_ids"].clone() # Add labels for Causal LM loss calculation
        return item

train_dataset = DiseaseDataset(train_encodings)
val_dataset = DiseaseDataset(val_encodings)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=1,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    eval_strategy="epoch", # Corrected from evaluation_strategy
    save_total_limit=1,
    logging_steps=10,
    report_to="none"
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer # Pass the tokenizer to the trainer
)
```

```python
trainer.train()
```

```python
trainer.save_model("./results")
tokenizer.save_pretrained("./results")
print("Model saved!")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model = AutoModelForCausalLM.from_pretrained("./results", device_map="auto")
tokenizer = AutoTokenizer.from_pretrained("./results")

pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
prompt = "Identify the disease based on symptoms.\nSymptoms: Fever, headache, body pain\n\nResponse:"
out = pipe(prompt, max_length=80)[0]["generated_text"]
print(out)
```

```python
# Generate predictions for a subset of the test data
import json
import pandas as pd
import re # Import the regular expression module

preds_data = []
rows_to_use = 20 # Number of rows to use for prediction and evaluation

with open("test.jsonl", "r") as f:
    for i in range(rows_to_use):
        line = f.readline()
        if not line:
            break
        item = json.loads(line)
        inp_symptoms = item["input"]

        # Extract actual disease from item["output"]
        actual_disease_match = re.search(r"Disease: ([^\n]+)", item["output"])
        actual_disease = actual_disease_match.group(1).strip() if actual_disease_match else "Unknown"

        prompt = f"Identify the disease.\nSymptoms: {inp_symptoms}\nResponse:"
        # Use max_new_tokens for better control over generated text length
        generated_output = pipe(prompt, max_new_tokens=50, do_sample=False)[0]["generated_text"]

        # Extract predicted disease from the generated output
        predicted_disease_match = re.search(r"Disease: ([^\n]+)", generated_output)
        predicted_disease = predicted_disease_match.group(1).strip() if predicted_disease_match else "No Disease Found"

        preds_data.append({"symptoms": inp_symptoms, "actual_disease": actual_disease, "predicted_disease": predicted_disease})

# Create a DataFrame with the actual and predicted diseases
df = pd.DataFrame(preds_data)
df.to_csv("predictions.csv", index=False)

print("Done! predictions.csv ready.")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# df is now the updated DataFrame from the previous cell with 'actual_disease' and 'predicted_disease' columns
df_simple = df.copy()

# Generate the confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix(df_simple["actual_disease"], df_simple["predicted_disease"]), annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted Disease")
plt.ylabel("Actual Disease")
plt.show()
```
